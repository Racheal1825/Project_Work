## Histogram with respect to Density
---

```python
bins = np.arange(0.990, 1.003, 0.001)

plt.figure(figsize = (7,5))
plt.hist(nData.density, bins = bins , rwidth = 0.7 )
plt.xticks(bins)

plt.title('Distribution of densities of Red Wine', fontdict  = {'fontweight': 'bold'})
plt.xlabel('Density')
plt.ylabel('Red wine')

plt.show()
```

## Scatter graph for Fixed acidity against pH
---

```python
%matplotlib inline
```

```python
plt.figure(figsize = (7,5))
plt.title('Fixed acidity against pH', fontdict = {'fontweight': 'bold'})
plt.scatter(x = 'pH', y = 'fixed_acidity', data = nData, c = 'b', alpha= 0.6)
plt.xlabel('pH')
plt.ylabel('Fixed acidity')

plt.legend()
```

###

## Pie Chart
---
#### The average amount of citric acid, residual sugar, chlorides , free sulfur dioxide, sulpates and alcohol in red wine

```python
nData_statistics[['citric_acid','residual_sugar','chlorides','free_sulfur_dioxide','sulphates', 'alcohol']].loc['mean']
```

```python
citric_acid = nData_statistics['citric_acid'].loc['mean']
residual_sugar = nData_statistics['residual_sugar'].loc['mean']
chlorides = nData_statistics['chlorides'].loc['mean']
free_sulfur_dioxide = nData_statistics['free_sulfur_dioxide'].loc['mean']
sulphates = nData_statistics['sulphates'].loc['mean']
alcohol = nData_statistics['alcohol'].loc['mean']

labels = ['Citric acid', 'Residual sugar','Chlorides','Free sulfur dioxide','Sulphates','Alcohol']
explode_values = [0.3,0,0.4,0,0.2,0]

plt.pie([citric_acid,residual_sugar,chlorides, free_sulfur_dioxide,sulphates,alcohol], labels = labels , autopct = '%.2f %%', explode = explode_values )
plt.title('Components of Red wine',fontdict = {'fontweight': 'bold'})
plt.show
```

##
